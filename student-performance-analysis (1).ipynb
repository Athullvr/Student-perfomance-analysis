```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

# 1. Load dataset (your correct path)
df = pd.read_csv("/kaggle/input/student-performance-and-teaching-dataset/Student_Performance_Teaching_Effectiveness.csv")

# 2. Basic info
df.info()
df.describe()

# 3. Check missing values
df.isnull().sum()

# 4. Visualizations

df.head()
# Gender distribution
df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.show()

# Attendance distribution
df['Attendance_Percentage'].plot(kind='hist')
plt.title('Attendance Percentage Distribution')
plt.show()

# Midterm Grade vs Final Grades
sns.regplot(x="Midterm_Grade",
y="Final_Grade",data=df)
plt.title("Midterm Grade vs Final Grade")
plt.show()

# 5. Correlation
df.corr(numeric_only=True)
df.head()

#6 Student Performance vs Study Hour Per Week
plt.figure(figsize=(8,5))
sns.boxplot(data=df,
x="Student_Performance",
y="Study_Hours_per_Week")
plt.title(" Study Hours Per Week vs `Student Performance")
plt.xlabel("Student Performance Level")
plt.ylabel(" Student Hours Per Week")
plt.show()
```
